# Decision Tree

集合の要素のばらつきが最も小さくなるように分割を繰り返す。

ばらつきの測り方はジニ不純度、エントロピー、分類誤差などがあるが、どれを使ってもあまり変わらないことが多い。

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append(".")
from modules import *
%matplotlib inline
plt.style.use('ggplot')
```

```python
# define impurity
def entropy(p):
    return -p*np.log2(p) - (1 - p)*np.log2((1-p))
def gini(p):
    return (p)*(1 - (p)) + (1 - p)*(1 - (1-p))
def error(p):
    return 1- np.max([p, 1 - p])

# make variables
x = np.arange(0.0, 1.0, 0.01)
ent = [entropy(p) if p != 0 else None for p in x]
sc_ent = [e*0.5 if e else None for e in ent] # why need to scale?
gini = gini(x)
err = [error(i) for i in x]

# plot
fig = plt.figure()
ax = plt.subplot(111)
for i, lab, ls, c, in zip([ent, sc_ent, gini, err],
                         ['Entropy', 'Entropy (scaled)', 'Gini Impurity', 'Misclassification Error'],
                         ['-', '-', '--', '-.'],
                         ['black', 'lightgray', 'red', 'green']):
    line = ax.plot(x, i, label=lab, linestyle=ls, lw=2, color=c)
    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y,  test_size = 0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
tree.fit(X_train, y_train)
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X_combined, y_combined, classifier=tree, test_idx=range(105,150))
plt.xlabel('petal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')
plt.show()
```

# Random Forest

各決定木を作るときにすべての特徴量を評価するのではないところがポイント。

チューニングをする必要はあまりなくて、決定木の個数だけ気にすれば大丈夫。

一般には多くすればそれだけ性能が良くなるが、計算コストが増える。

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion='entropy', n_estimators=10, random_state=1, n_jobs=2)

forest.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined, classifier=forest, test_idx=range(105,150))
plt.xlabel('petal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')
plt.show()
```
