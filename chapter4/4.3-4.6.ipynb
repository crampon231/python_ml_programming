# 4.3 training data and test data

```python
import pandas as pd
import numpy as np
```

```python
df_wine = pd.read_csv('wine.data', header=None)
df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
                   'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                   'Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
df_wine.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = df_wine.iloc[:, 1:], df_wine.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train
```

# 4.4 Scaling

```python
# min-max scaling
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.transform(X_test)
pd.DataFrame(X_train_norm).describe()
```

æ¨™æº–åŒ–ã®æ–¹ãŒå®Ÿç”¨çš„ã‹ã‚‚ã€‚å¤šãã®ç·šå½¢ãƒ¢ãƒ‡ãƒ«ãŒé‡ã¿ã‚’0ã«è¿‘ã„ä¹±æ•°ã§åˆæœŸåŒ–ã™ã‚‹ã‹ã‚‰ã€‚æ¨™æº–åŒ–ã¯å¤–ã‚Œå€¤ã«å¯¾ã—ã¦ã‚‚å¼·ã„ã€‚

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
pd.DataFrame(X_train_std).describe()
```

# 4.5 feature selecting

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty='l1', C=0.1)
lr.fit(X_train_std, y_train)
print('Training accuracy:', lr.score(X_train_std, y_train))
print('Test accuracy:', lr.score(X_test_std, y_test))
```

```python
# No scaling
lr_ns = LogisticRegression(penalty='l1', C=0.1)
lr_ns.fit(X_train, y_train)
print('Training accuracy:', lr_ns.score(X_train, y_train))
print('Test accuracy:', lr_ns.score(X_test, y_test))
```

regularization path

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

weights, params= [], []
for c in np.arange(-4, 6):
    lr = LogisticRegression(penalty='l1', C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10**c)
    
pd.DataFrame(weights, columns= df_wine.columns[1:], index=params).plot(figsize=(10,8), logx=True)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2)
sbs = SFS(knn, 
          k_features=3, 
          forward=False, 
          floating=False, 
          scoring='accuracy',
          cv=4,
          n_jobs=-1)
sbs.fit(X_train_std, y_train)
```

```python
pd.DataFrame.from_dict(sbs.get_metric_dict()).T
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
fig = plot_sfs(sbs.get_metric_dict(), kind='std_dev')
fig.set_size_inches((10,8))
plt.ylim([0.8, 1])
plt.title('Sequential Backward Selection (w. StdDev)')
plt.grid()
plt.show()
```

```python
best_k = pd.DataFrame.from_dict(sbs.get_metric_dict()).T['avg_score'].argmax()
best_features = list(pd.DataFrame.from_dict(sbs.get_metric_dict()).T.ix[best_k, 'feature_idx'])
df_wine.columns[1:][best_features]
```

```python
knn.fit(X_train_std, y_train)
print('Training accuracy:', knn.score(X_train_std, y_train))
print('Test accuracy:', knn.score(X_test_std, y_test))
```

```python
knn.fit(X_train_std[:, best_features], y_train)
print('Training accuracy:', knn.score(X_train_std[:, best_features], y_train))
print('Test accuracy:', knn.score(X_test_std[:, best_features], y_test))
```

# 4.6 Random Forest to access importance of features

```python
from sklearn.ensemble import RandomForestClassifier
feat_labels = df_wine.columns[1:]
forest = RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1)
forest.fit(X_train, y_train)
```

```python
importance = pd.DataFrame(forest.feature_importances_)
importance.index = feat_labels
importance.sort_values(0, ascending=False).plot(kind='bar', legend=False)
```
